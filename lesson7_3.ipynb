```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager
import matplotlib as mlp
fontManager.addfont('ChineseFont.ttf')
mlp.rc('font',family='ChineseFont')
```

```python
figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot(1,1,1)
axes.plot([1,3,5,7,9],marker='o',color='red',linestyle='--')


plt.show()
```

```python
figure = plt.figure(figsize=(10,6))
ax1 = figure.add_subplot(2,1,1)
ax1.plot([1,3,5,7,9],marker='o',color='red',linestyle='-')
ax2 = figure.add_subplot(2,1,2)
ax2.plot([1,3,5,7,15],marker='o',color='green',linestyle='-.')
ax2.bar(['a', 'b', 'c', 'd', 'e'],[1, 2, 3, 4, 5])

plt.show()
```

```python
student1 = {'':30, '':80, '':70, '':90, '':50}
student2 = {'':50, '':60, '':30, '':80, '':60}
```

```python
subjects = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize=(10,8))
axes = figure.add_subplot(1,1,1)
axes.set_title('')
axes.set_xlabel('')
axes.set_ylabel('')
axes.plot(subjects,score1,'ro-.',label='')
axes.plot(subjects,score2,'gx--',label='')
axes.legend()
axes.grid(axis='y')

plt.show()
```
